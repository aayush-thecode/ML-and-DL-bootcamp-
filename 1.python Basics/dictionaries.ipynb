```python
## intoduction to dictionaries 

## Dictionaries are unordered collections of items. They store data in key-value pairs. Keys must be unique and immutable (e.g. strings, numbers or tupples), while values can be of any type.
```

```python
## creating Dictionaries 
empty_dict = {}
print(type(empty_dict))
```

```python
empty_dict = dict()
empty_dict
```

```python
student = {"name":"Aayush","age": 32,"grade":24}
print(student)
print(type(student))
```

```python
# Error / single key is always used 

student = {"name":"Aayush","age": 32,"grade":24}
print(student)
```

```python
## accessing dictionary elements

student = {"name":"Aayush","age": 32,"grade":'A'}
print(student)
```

```python
## Accessing Dictionary elements
print(student['grade'])
print(student['age'])

## Accessing using get() method
print(student.get('grade'))
print(student.get('last_name'))

print(student)
```

```python
## modifying dicitonary elements
## dictionary are mutable, so you can add, update or delete elements

student["age"] = 33 #upated value and key
print(student)
student["address"] = "Nepal" ##added a new key and value
print(student)
```

```python
del student['grade'] ## delete key and value pair

print(student)
```

```python
## Dictionary methods

keys = student.keys()
print(keys)
values = student.values()
print(values)

items = student.items() ## get all key value pairs
print(items)
```

```python
## shallow copy 

student_copy = student
print(student)
print(student_copy)
```

```python
student['name'] = 'Aayush 1'
print(student)
print(student_copy)
```

```python
student_copy1 = student.copy() ## shallow copy
print(student_copy1)
print(student)
```

```python
student['name'] = 'aayush2'
print(student)
print(student_copy1)
```

```python
### Iterating over Dictionaries
## you can use loops to iterate over dictionaries, keys, values or items

## Iterating over keys
for keys in student.keys():
    print(keys)
```

```python
## Iterate over values 

for value in student.values():
    print(value)
```

```python
## iterate over key value pairs 

for key, value in student.items():
    print(f'{key}:{value}')
```

```python
## Nested Dictionaries
students = {
    "student1":{"name":"Aayush", "age": 22},
    "student2":{"name":"Samuel", "age": 23}
}
print(students)
```

```python
##accessing nested dictionaries elements
print(students['student2']['name'])
print(students['student2']['age'])
```
