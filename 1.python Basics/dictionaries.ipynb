```python
## intoduction to dictionaries 

## Dictionaries are unordered collections of items. They store data in key-value pairs. Keys must be unique and immutable (e.g. strings, numbers or tupples), while values can be of any type.
```

```python
## creating Dictionaries 
empty_dict = {}
print(type(empty_dict))
```

```python
empty_dict = dict()
empty_dict
```

```python
student = {"name":"Aayush","age": 32,"grade":24}
print(student)
print(type(student))
```

```python
## accessing dictionary elements

student = {"name":"Aayush","age": 32,"grade":'A'}
print(student)
```

```python
## Accessing Dictionary elements
print(student['grade'])
print(student['age'])

## Accessing using get() method
print(student.get('grade'))
print(student.get('last_name'))
```
