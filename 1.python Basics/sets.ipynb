SET

Sets are built-in data type in python used to store collections of unique items. They are unordered, meaning that the elements do not follow a specific order, and do not allow duplicate elements. Sets are useful for membership tests, eliminating duplicate entries and performing mathematical sets like union, intersection, difference and symmetric difference 

```python
## creating a set
my_set = {1,2,3,4,5}
print(my_set)
print(type(my_set))
```

```python
my_set = set([1,2,3,4,5,6])
print(type(my_set))
```

```python
## basics sets operation
## adding and removing elements

my_set.add(7)
print(my_set)

my_set.remove(7)
print(my_set)
```

```python
my_set.discard(1)
print(my_set)
```

```python
## pop method
my_set = ([1,2,3,4,5])

removed_element = my_set.pop()

print(removed_element)
print(my_set)
```

```python
## clear all the elements

my_set = [(1,2,3,4,5)]

my_set.clear()
print(my_set)
```

```python
## set membership test

my_set = {1,2,3,4,5}
print(3 in my_set)
print(10 in my_set)
```

```python
## mathematical operation

set1 = {1,2,3,4,5,6}
set2 = {4,5,6,7,8,9}

## union 
union_set = set1.union(set2)
print(union_set)

## intersection
intersection_set = set1.intersection(set2)
print(intersection_set)

set1.intersection_update(set2)
print(set1)
```

```python
set1 = {1,2,3,4,5,6}
set2 = {4,5,6,7,8,9}

##difference
print(set1.difference(set2))


## symmetric difference
set1.symmetric_difference(set2)
```

```python
## sets methods

set1 = {1,2,3,4,5}
set2 = {3,4,5}

##is subset
print(set1.issubset(set2))

print(set1.issuperset(set2))
```

```python
lst = [1,2,2,3,4,4,5]

set(lst)
```

```python
## counting unique words in text

text = "we are discussing about sets"
words = text.split()

## convert list of words to set to get unique words 

unique_words = set(words)
print(unique_words)
```
