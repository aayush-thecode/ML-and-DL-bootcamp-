```python
### variable length arguements
## positional and keywords arguements

def print_numbers(*args):
    for numbers in args:
        print(numbers)
```

```python
### keywords Arguements

def print_details(**kwargs):
    for key, value in kwargs.items():
        print(f'{keys}:{value}')
```
